## Categorical Plots:

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(citi_clean['trip_duration_minutes'], bins = 20, kde = True)
```

The majority of trips is between 7 and 16 minutes. 

```python
citi_clean['trip_duration_minutes'].describe()
```

The average trip duration is 13 minutes. The median trip length is 10 minutes. The distribution is right-skewed. There is another peak at the end of the longtail, which could be longer trips like day trips accumulated. 

The shortest trip was one minute, the longest trip was 45 minutes. As there is the aggregation at the 45 minute mark, there seems to be an accumulation (trips longer than 45 minutes are cut off at the 45minute mark).

It could be that only short trips are counted in the data, and longer trips are not available or cut off.

```python
citi_clean.loc[citi_clean['trip_duration_minutes'] < 7, 'Duration category'] = 'Short Duration'
```

```python
citi_clean.loc[(citi_clean['trip_duration_minutes'] >= 7) & (citi_clean['trip_duration_minutes'] < 16), 'Duration category'] = 'Medium Duration'
```

### Trip data

```python
citi_clean['end_time'].max()
```

```python
citi_clean['start_time'].min()
```

The first trip in the data was recorded on September 01, 2013. The last trip was finished on October 1st, 2013. So we only have trip data for one month.

```python
citi_clean['start_time'].describe()
```

```python
# Convert the "start_time" column to a datetime type
citi_clean['start_time'] = pd.to_datetime(citi_clean['start_time'])

# Extract the date from the "start_time" column and create a new column "start_date"
citi_clean['start_date'] = citi_clean['start_time'].dt.date
```

```python
# Group the data by "start_date" and count the occurrences of each date
count_by_date = citi_clean.groupby('start_date')['trip_id'].count()

# Create the time series chart
plt.plot(count_by_date.index, count_by_date.values)

# Set the plot title and axes labels
plt.title('Trip Count Time Series')
plt.xlabel('Start Date')
plt.ylabel('Trip Count')

# Rotate the x-axis tick labels for better readability (optional)
plt.xticks(rotation=45)

# Save the plot as an image file 
plt.savefig('trip_count_time_series.png')

# Display the time series chart
plt.show()
```

The weekend to weekday pattern is clearly visible. THe overall trend in trip volume is increasing.

## Hypotheses

1. The overall trip volume is increasing. 
2. The trip volume is higher during the week than on the weekend.
3. The service attracts younger customers (<50) than elderly customers.
